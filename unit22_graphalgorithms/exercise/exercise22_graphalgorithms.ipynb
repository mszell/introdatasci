## 22.5
Apply Prim's and Kruskal's algorithm to the graph shown in the figure by hand. Then use the code below to verify your answer.




```python
edgelist = [[1,2, {'weight': 10}],
            [2,3,{'weight': 2}],
            [3,4, {'weight':8}],
            [4,5, {'weight': 3}],
            [5,6, {'weight': 11}],
            [6,1, {'weight': 9}],
            [1,7, {'weight': 1}],
            [2,7, {'weight': 6}],
            [3,7, {'weight': 5}],
            [4,7, {'weight': 7}],
            [5,7, {'weight': 4}],
            [6,7, {'weight': 12}],
           ]
    
G2 = nx.from_edgelist(edgelist)
```

```python
G_mst = nx.from_edgelist(nx.minimum_spanning_edges(G2, algorithm='kruskal'))
```

```python
nx.draw_networkx(G_mst)
```
